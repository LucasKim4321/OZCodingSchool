## 3. while문 조건식 증감, 감소

```python
#1. 기본적인 증가 유형

i = 0 
while i < 5:
    print('점점 성장하는 여러분!',i)
    i += 1
```

```python
#2. 기본적인 감소 유형

i = 5 
while i > 0:
    print('점점 성장하는 여러분!',i)
    i -= 1
```

## 5. while문 for문보다 강점인 부분

```python
# 조건을 만족할 때까지 무한으로 실행하기 때문에 코드가 간편해질 수 있다.
# 무수히 많이 실행할 때 사용.

import random

random.randint(1,6)

i = 0
while i != 4:
    i = random.randint(1,6)
    print(i)
```

## 6. continue, break

```python
# i = 0
# while True:
#     print(i)
#     i += 1
#     if i == 5:
#         break  # 중단하기

i = 0
for i in range(100):
    print(i)
    if i == 5:
        break  # 중단하기

```

```python
i = 0
while i < 5:
    i += 1
    if i % 2 == 0:
        continue  #건너뛰기
    print(i)

i = 0
for i in range(6):
    if i % 2 == 0:
        continue  #건너뛰기
    print(i)
```


## 7. while문에도 else가 있어요!

```python
# else : while문을 정상 종료 후 실행
# 조건식을 만족해 종료시 실행
# break로 종료시 실행 안됨

양파 = 5
while 양파 > 0:
    print("양파", 양파, "개 남았습니다.")
    if 양파 <= 0:
        break
    양파 -= 1
else:
    print("양파 다 팔렸습니다.")
```

## 8. 붕어빵 인터네셔널은 내가 살린다 나도 이제 개발자

```python
# dictionary를 이용해 개선
주문_개수 = map(int, input(f'붕어빵 맛 순서대로 개수를 입력해주세요. 안살꺼면 0입력(팥, 슈크림, 잡채 순으로 입력)(각 개수를 입력하고 공백으로 구분하세요.)').split(' '))

품목별_가격 = {"팥":2000, "슈크림":2500,"잡채":3000}

주문_개수 = list(주문_개수)
총_주문_개수 = sum(주문_개수)
total = 0

total += 품목별_가격['팥'] * 주문_개수[0]
total += 품목별_가격['슈크림'] * 주문_개수[1]
total += 품목별_가격['잡채'] * 주문_개수[2]

#출력
print(f"주문이 완료되었습니다. 팥 {주문_개수[0]}개, 슈크림 {주문_개수[1]}개, 잡채 {주문_개수[2]}개 총 {총_주문_개수}개 결제 금액은 {total}원 입니다.")
```

## 9. 예제로 while문 익혀보기
![스크린샷 2025-01-07 144415.png](<attachment:스크린샷 2025-01-07 144415.png>)

```python
num = 4
print("초기값:",num)
count= 0

while num > 1:
    if num % 2 != 0:  #홀수
        num = num * 3 + 1
    else:  #짝수
        num = num // 2
    print(num)
    count += 1
else:
    print(f'{count}회 반복하였습니다!')
```
