## 1. 객체

객체 : 구현하고자 하는 대상의 **속성과 기능**을 가지는 프로그램 단위

ex) 사람을 객체로 만들어봅시다.

- 속성 : 눈, 코, 입, 키, 몸무게, 성격 등
- 기능 : 다리를 이용해 걷기, 물건을 잡을 때 손가락을 이용하기

객체를 만들기 위해서는 속성과 기능이 필요합니다.

하지만, 현실 세계에 있는 물건이든, 개념 또는 사상 등을 많은 것들을 프로그램으로 표현하기에는 대상들이 가지고 있는 구성 요소 또는 특징 그리고 기능들이 많습니다.

⇒ 객체를 구성하고 있는 속성이 많은 경우, 필요한 속성만을 이용해 객체를 표현하는 개념을 추상화라고 합니다.

ex) 개발 공부를 하는 수강생을 객체로 만들기 위해서 필요한 속성과 기능은 무엇일까요?

- 이름
- Python 활용 능력
- Database 활용 능력
- Django 활용 능력
- AWS 활용 능력

생각할 수 있는 기능은 무엇이 있을까요?

- 수강 과목별 능력 평균 구하기
- 수강 과목별 활용 능력 총점 구하기

````python
# oz_students 리스트는 각 학생의 이름과 그들의 Python, Database, Django, AWS 능력 점수를 저장한 딕셔너리들의 리스트입니다.
oz_students = [
    {'name':'백현우', 'Python':4, 'Database':3, 'Django':3, 'AWS':2},
    {'name':'홍혜인', 'Python':4, 'Database':5, 'Django':2, 'AWS':4},
    {'name':'윤은성', 'Python':3, 'Database':4, 'Django':4, 'AWS':1},
    {'name':'홍수철', 'Python':2, 'Database':3, 'Django':1, 'AWS':5},  # python에서는 마지막에 콤머(,) 찍어도 되는가봄. 자바에선 에러남
]

print('이름','총점','평균', sep = '\t')

# oz_students 리스트의 각 학생에 대해 반복문을 실행합니다.
for student in oz_students:
    ability_sum = student['Python'] + student['Database'] + student['Django'] + student['AWS']
    ability_average = ability_sum / len(oz_students)

    print(student['name'], ability_sum, ability_average, sep ='\t')

oz_students = [
    {"name": "백현우", "python": 4, "database": 3, "Django": 3, "AWS": 2},
    {"name": "홍혜인", "python": 4, "database": 5, "Django": 2, "AWS": 4},
    {"name": "윤은성", "python": 3, "database": 4, "Django": 4, "AWS": 1},
    {"name": "홍수철", "python": 2, "database": 3, "Django": 1, "AWS": 5},
]

# 출력할 표의 헤더를 설정합니다. 이름, 총점, 평균을 탭(\t)으로 구분하여 출력합니다.
print("이름", "총점", "평균", sep="\t")
````

여러가지 속성(이름, 수강한 과목) 등에 대한 **정보**를 가지고 있는 대상을 객체라고 합니다.

프로그래밍 언어에서 좀 더 넓은 의미로 **속성을 가질 수 있는 모든 것**들을 객체라고 합니다. 

## 2. 함수를 이용한 객체 사용

````python
# create_oz_students 함수는 학생의 이름과 각 과목 점수를 입력받아 학생 정보를 딕셔너리 형태로 반환합니다.
def create_oz_students(name, python, database, django, AWS):
    return {
        'name' : name,
        'Python' : python,
        'Database' : database,
        'Django' : django,
        'AWS' : AWS
    }

# oz_students 리스트는 create_oz_students 함수를 사용하여 각 학생의 정보를 포함하는 딕셔너리들의 리스트를 생성합니다.
# name, python, database, django, AWS
oz_students = [
    create_oz_students('백현우',4,3,3,2),
    create_oz_students('홍혜인',4,5,2,4),
    create_oz_students('윤은성',3,4,4,1),
    create_oz_students('홍수철',2,3,1,5)
]

print('이름','총점','평균', sep = '\t')

for student in oz_students:
    ability_sum = student['Python'] + student['Database'] + student['Django'] + student['AWS']
    ability_average = ability_sum / len(oz_students)

    print(student['name'], ability_sum, ability_average, sep ='\t')
````

````python
def create_oz_students(name, python, database, django, AWS):
    return {
        'name' : name,
        'Python' : python,
        'Database' : database,
        'Django' : django,
        'AWS' : AWS
    }

def get_sum(student):
    return student['Python'] + student['Database'] + student['Django'] + student['AWS']

def get_average(student):
    return get_sum(student) / len(oz_students)

def to_string(student):
    return f"{student['name']}\t{get_sum(student)}\t{get_average(student)}"

oz_students = [
    create_oz_students('백현우',4,3,3,2),
    create_oz_students('홍혜인',4,5,2,4),
    create_oz_students('윤은성',3,4,4,1),
    create_oz_students('홍수철',2,3,1,5)
]

print(oz_students)
print(oz_students[1]['name'])
print('이름', '총점', '평균', sep = '\t')

for student in oz_students:
    print(to_string(student))
````


**객체와 관련 코드를 분리**할 수 있게 하는 것이 객체 지향 프로그램의 핵심입니다.

개발자들은 이러한 객체 지향 프로그램을 반복적으로 만들다보니 이러한 구조를 자동으로 만들어줄 클래스 구조를 만들게 되었습니다.

```python
	클래스 내용
```

1. 객체란 무엇인가요?
    - 정답
        
        객체는 구현하고자 하는 대상의 **속성과 기능을 가지는 프로그램 단위**입니다. 객체는 속성과 기능을 통해 실제 세계의 사물이나 개념을 모델링한 것입니다. 예를 들어, 사람 객체는 속성으로 이름, 나이, 키 등을 가질 수 있고, 기능으로 걷기, 말하기 등을 가질 수 있습니다.
        
    
2. 객체를 구성하는 속성과 기능에 대해 설명해주세요.
    - 정답
        
        객체를 구성하는 속성은 객체가 가지는 **데이터**를 나타내며, 기능은 객체가 수행할 수 있는 **동작**을 나타냅니다. 예를 들어, 사람 객체의 속성으로는 이름, 나이, 키 등이 있고, 기능으로는 걷기, 말하기, 먹기 등이 있습니다.
        
    
3. 추상화에 대해 설명해주세요.
    - 정답
        
        추상화는 현실 세계의 복잡한 객체를 프로그램으로 표현할 때, 필요한 **속성과 기능만을 선택하여 모델링**하는 과정입니다. 불필요한 세부 사항을 제거하고 중요한 특성만을 남기는 것을 의미합니다. 예를 들어, 학생 객체를 만들 때 모든 신체적 특성을 포함하지 않고, 이름, 성적 등 필요한 정보만 포함하는 것입니다.
        
    
4. 객체 지향 프로그램에 대해 설명해주세요.
    - 정답
        
        객체 지향 프로그램은 **객체를 중심으로 프로그램을 설계하고 구현**하는 방식입니다. 객체 지향 프로그래밍(OOP)의 핵심 개념으로는 캡슐화, 상속, 다형성, 추상화가 있습니다. OOP는 코드의 재사용성과 유지보수성을 높이며, 복잡한 시스템을 더 쉽게 관리할 수 있게 합니다.
        
    
5. 클래스를 선언하는 구조에 대해 설명해주세요.
    - 정답
        
        클래스는 객체를 생성하기 위한 청사진 또는 템플릿입니다. 클래스를 선언하는 구조는 다음과 같습니다:
        
        ```python
        class 클래스이름:
            클래스 내용
        ```
        
        클래스 이름은 대문자로 시작하는 것이 관례이며, 클래스 내부에는 속성(변수)와 기능(메서드)을 정의합니다. 예를 들어:
        
        ```python
        python코드 복사
        class Person:
            def __init__(self, name, age):
                self.name = name
                self.age = age
        
            def greet(self):
                print(f"안녕하세요 저의 이름은 {self.name}이고, 나이는 {self.age}살 입니다.")
        
        ```
        
    
6. 인스턴스에 대해 설명해주세요.
    - 정답
        
        인스턴스는 **클래스를 사용하여 생성된 객체**를 의미합니다. 인스턴스는 클래스의 구체적인 구현체로, 클래스의 속성과 기능을 실제로 사용할 수 있게 합니다. 인스턴스를 생성하려면 클래스 이름을 함수처럼 호출합니다. 예를 들어:
        
        ```python
        python코드 복사
        person1 = Person("Alice", 30)
        person2 = Person("Bob", 25)
        
        person1.greet()  # 출력: Hello, my name is Alice and I am 30 years old.
        person2.greet()  # 출력: Hello, my name is Bob and I am 25 years old.
        
        ```
        
        여기서 person1과 person2는 Person 클래스의 인스턴스입니다.
