1. 객체란 무엇인가요?
    - 정답
        
        객체는 구현하고자 하는 대상의 **속성과 기능을 가지는 프로그램 단위**입니다. 객체는 속성과 기능을 통해 실제 세계의 사물이나 개념을 모델링한 것입니다. 예를 들어, 사람 객체는 속성으로 이름, 나이, 키 등을 가질 수 있고, 기능으로 걷기, 말하기 등을 가질 수 있습니다.
        
    
2. 객체를 구성하는 속성과 기능에 대해 설명해주세요.
    - 정답
        
        객체를 구성하는 속성은 객체가 가지는 **데이터**를 나타내며, 기능은 객체가 수행할 수 있는 **동작**을 나타냅니다. 예를 들어, 사람 객체의 속성으로는 이름, 나이, 키 등이 있고, 기능으로는 걷기, 말하기, 먹기 등이 있습니다.
        
    
3. 추상화에 대해 설명해주세요.
    - 정답
        
        추상화는 현실 세계의 복잡한 객체를 프로그램으로 표현할 때, 필요한 **속성과 기능만을 선택하여 모델링**하는 과정입니다. 불필요한 세부 사항을 제거하고 중요한 특성만을 남기는 것을 의미합니다. 예를 들어, 학생 객체를 만들 때 모든 신체적 특성을 포함하지 않고, 이름, 성적 등 필요한 정보만 포함하는 것입니다.
        
    
4. 객체 지향 프로그램에 대해 설명해주세요.
    - 정답
        
        객체 지향 프로그램은 **객체를 중심으로 프로그램을 설계하고 구현**하는 방식입니다. 객체 지향 프로그래밍(OOP)의 핵심 개념으로는 캡슐화, 상속, 다형성, 추상화가 있습니다. OOP는 코드의 재사용성과 유지보수성을 높이며, 복잡한 시스템을 더 쉽게 관리할 수 있게 합니다.
        
    
5. 클래스를 선언하는 구조에 대해 설명해주세요.
    - 정답
        
        클래스는 객체를 생성하기 위한 청사진 또는 템플릿입니다. 클래스를 선언하는 구조는 다음과 같습니다:
        
        ```python
        class 클래스이름:
            클래스 내용
        ```
        
        클래스 이름은 대문자로 시작하는 것이 관례이며, 클래스 내부에는 속성(변수)와 기능(메서드)을 정의합니다. 예를 들어:
        
        ```python
        python코드 복사
        class Person:
            def __init__(self, name, age):
                self.name = name
                self.age = age
        
            def greet(self):
                print(f"안녕하세요 저의 이름은 {self.name}이고, 나이는 {self.age}살 입니다.")
        
        ```
        
    
6. 인스턴스에 대해 설명해주세요.
    - 정답
        
        인스턴스는 **클래스를 사용하여 생성된 객체**를 의미합니다. 인스턴스는 클래스의 구체적인 구현체로, 클래스의 속성과 기능을 실제로 사용할 수 있게 합니다. 인스턴스를 생성하려면 클래스 이름을 함수처럼 호출합니다. 예를 들어:
        
        ```python
        python코드 복사
        person1 = Person("Alice", 30)
        person2 = Person("Bob", 25)
        
        person1.greet()  # 출력: Hello, my name is Alice and I am 30 years old.
        person2.greet()  # 출력: Hello, my name is Bob and I am 25 years old.
        
        ```
        
        여기서 person1과 person2는 Person 클래스의 인스턴스입니다.
