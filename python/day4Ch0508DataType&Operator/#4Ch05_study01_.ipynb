##변수를 만드는 규칙

영문 문자와 숫자를 사용할 수 있습니다.

대소문자를 구분합니다.

첫글자는 대문자로 쓰지 않습니다.

문자부터 시작해야 하며 숫자부터 시작하면 안됩니다.

_(밑줄 문자)로 시작할 수 있습니다.

특수 문자( +, -, *, /, $, @, &, % 등)는 사용할 수 없습니다.

파이썬의 키워드( if, for, while, and, or 등)는 사용할 수 없습니다.

```python
# 대소문자를 구분하는 변수
BIG = "크다"
big = "작다"
print(BIG,big)
```
