```python
# 기본적으로 함수안에서 전역 변수를 수정하려면 특별한 방법을 써야함.
# 'global 변수이름' 사용

a = 100
def func1():
    global a  # 파이썬에서 전역 변수를 불러오려면 이렇게 불러와야함.
    a += 1
    print(a)
    return a

func1()
```

```python
# dict 타입은 update()를 사용하면 global 설정없이 수정도 됨.
# 메소드를 통해 접근하기 때문인거 같음. java의 setter,getter이랑 같은 이치

won = {"A등급":1000, "B등급":500}

def 계산 (x, y):
    z =  x * won['A등급'] + y * won['B등급']
    won.update({"C등급":2000})
    return z

print(계산(10,2),won)
```

```python
# 1. LEGB 규칙이란?
# Python에서 변수의 범위를 찾는 순서는 다음과 같습니다:

# L (Local, 지역 범위)

# 함수 또는 블록 안에서 정의된 변수입니다.
# 가장 먼저 찾습니다.
# E (Enclosing, 비지역 범위)

# 중첩된 함수의 바깥 함수에서 정의된 변수입니다.
# nonlocal 키워드로 접근할 수 있습니다.
# G (Global, 전역 범위)

# 함수 밖에서 정의된 변수입니다.
# global 키워드로 접근할 수 있습니다.
# B (Built-in, 내장 범위)

# Python의 내장 함수나 변수입니다.
# print, len, sum 등이 이에 해당합니다.
```

```python
# 함수 안에서 읽어 오는건 global여부와 상관 없음.
# 수정할 때 global 사용.

A = 1
B = 2
C = 3
유물 = 0

def 감자():
    global A
    A += 1
    # print(A)

def 양파():
    global A, B, C
    A += 3
    B += 3
    C += 4

def 옥수수():
    global 유물
    유물 += 1

감자()
양파()
옥수수()

합계 = A + B + C + 유물
print('오늘 수확한 수확물 :', 합계)
```

```python

def 수확철():
    야채 = {'A등급': 0, 'B등급': 0, 'C등급': 0}

    def 양파():
        야채['A등급'] += 1

    def 대파():
        야채['A등급'] += 3
        야채['B등급'] += 3
        야채['C등급'] += 4

    def 쪽파():
        야채['금'] = 1

    양파()
    대파()
    쪽파()

    return 야채

# 함수 호출 및 결과 출력
print(수확철())
print(sum(수확철().values()))
```

| **1** | `(값...)`, `[값...]`, `{키:값...}`, `{값...}` | **튜플, 리스트, 딕셔너리, 세트 생성** |
| **2** | `x[인덱스]`, `x[인덱스:인덱스]`, `x(인수...)`, `x 속성` | **리스트(튜플) 첨자, 슬라이싱, 함수 호출, 속성 참조** |
| **3** | `await x` | **await 표현식** |
| **4** | `**` | **거듭제곱** |
| **5** | `+x`, `-x`, `~x` | **단항 덧셈(양의 부호), 단항 뺄셈(음의 부호), 비트 NOT** |
| **6** | `*`, `@`, `/`, `//`, `%` | **곱셈, 행렬 곱셈, 나눗셈, 버림 나눗셈, 나머지** |
| **7** | `+`, `-` | **덧셈, 뺄셈** |
| **8** | `<<`, `>>` | **비트 시프트** |
| **9** | `&` | **비트 AND** |
| **10** | `^` | **비트 XOR** |
| **11** | `\|` | **비트 OR** |
| **12** | `in`, `not in`, `is`, `is not`, `<`, `<=`, `>`, `>=`, `==`, `!=` | **포함 연산자, 객체 비교 연산자, 비교 연산자** |
| **13** | `not x` | **논리 NOT** |
| **14** | `and` | **논리 AND** |
| **15** | `or` | **논리 OR** |
