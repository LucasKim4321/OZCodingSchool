```python
# class 클래스 이름:
#     def 매서드(self):
#         코드

class Person:
    def hello(self):
        print('수강생 여러분')

호출 = Person()  #인스턴스 호출
호출.hello()  #메서드 호출

# 메서드
# .index() .count()
```

```python
a = int(10) #a인스턴스
print(a)

b = list(range(10)) #b인스턴스
print(b)

# 인스턴스와 객체 무슨 차이? 사실은 인스턴스와 객체는 같아요.
# 인스턴스 + 클래스 인스턴스  /  객체 자체를 언급할땐 객체
```

```python
class Person:
    def __init__(self):
        self.hello = "안녕하세요"
    
    def greeting(self):
        print(self.hello)

호출 = Person()
호출.greeting()
```

그런데 도대체 self는 뭘까요?

self는 인스턴스 자기 자신을 의미합니다.

여기서 __init__의 매개변수 self에 들어가는 값은 Person()이라고 할 수 있다.


```python
# class 클래스이름:
#     def __init__(self, 매개변수1, 매개변수2):
#         self.속성1 = 매개변수1
#         self.속성2 = 매개변수2
```

```python
class Person:
    def __init__(self, name, age, address):
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.address = address
    
    def greeting(self):
        print("{0} 저는 {1}입니다." .format(self.hello, self.name))

maria = Person('마리아', 20, '서울시 서초구 반포동')
maria.greeting()   # 안녕하세요. 저는 마리아입니다.

print('이름:', maria.name)  # 마리아
print('나이:', maria.age)  # 20
print('주소:', maria.address)  # 서울시 서초구 반포동
```

## 클래스의 상속
![스크린샷 2025-01-08 214109.png](<attachment:스크린샷 2025-01-08 214109.png>)

```python
# 클래스 상속 : 다른 클래스의 기능을 상속 받는 것것
# class 클래스이름(상속받을 클래스이름)

class Car(object):
    maxspped = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):
    battery = 1000
    batteryKM = 300
    
k5 = HybridCar()
k5.maxspped  #300이 나옴

# print(dir(k5))
# print(dir(object))
```

```python
# 상속

class Car(object):
    maxspped = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class ElectricCar(HybridCar):
    battery = 2000
    batteryKM = 300
    
    
k5 = HybridCar()
# k5.maxspped  #300이 나옴

electricCarK5 = ElectricCar()
electricCarK5.maxspped  #300이 나옴
electricCarK5.battery  #2000이 나옴 #오버라이딩

# print(dir(k5))
# print(dir(object))
```

```python
# 메서드 오버라이딩

class Car(object):
    maxspped = 300
    maxpeople = 5
    def move(self, x):
        print("스피드로 달리고 있습니다.")
    def stop(self):
        print("멈췄습니다.")

class HybridCar(Car):
    battery = 1000
    batteryKM = 300

class ElectricCar(HybridCar):
    battery = 2000
    # batteryKM = 500
    def move(self, x):
        print(self.batteryKM, "만큼 달릴 수 있습니다.")
        print(x, "스피드로 달리고 있습니다.")
    
    
k5 = HybridCar()
# k5.maxspped  #300이 나옴

electricCarK5 = ElectricCar()
electricCarK5.maxspped  #300이 나옴
electricCarK5.battery  #2000이 나옴 #오버라이딩
electricCarK5.move(10)

# print(dir(k5))
# print(dir(object))
```
