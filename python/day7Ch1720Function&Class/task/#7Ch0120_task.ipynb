__[과제 3: 주문 처리 메서드 process_order]__

목표: process_order 함수를 구현하여 손님의 주문을 처리하고 재고를 업데이트합니다.

설명: bread_type과 bread_count를 매개변수로 받아, 재고를 확인하고 충분한 경우 재고를 감소시키고 판매량을 증가시킵니다. 재고가 부족하거나 잘못된 붕어빵 종류인 경우 적절한 메시지를 출력하세요.

```python
def process_order(self):
        # 매 주문 마다 주문 초기화
        order = {}

        while True:
            fish_type = input("주문할 붕어빵의 종류를 입력하세요 1.팥 2.슈크림 3.초코 4.주문 종료 (중복 주문시 기존 주문을 덮어씌움)").strip()
            # 숫자 입력시에만 형변환
            if not(fish_type.isdigit()):
                print("숫자만 입력해주세요!")
                continue
            else:
                fish_type = int(fish_type)
                
            # 주문 기능 범위를 벗어나면 경고
            #if의 조건문에서 형변환이 계속 유지됨. int()형변환 뒤 문자형 숫자인지 검사하면 문자가 아니라고 에러가 남! 조건식만 그럼. 어이 x
            if fish_type > (len(self.fish)+1) or fish_type < 1:
                # print(type(fish_type))
                print("1부터 4까지의 숫자만 입력해주세요! :b")
                continue

            # 숫자 4 입력시 주문 종료
            if fish_type == 4:
                amount_sum = 0
                total_payment = 0
                for i in order:
                    self.stocks.update({ i : (self.stocks[i] - order[i]) })  # 재고 업데이트
                    self.sold_fish.update({ i : (self.sold_fish[i] + order[i])})  # 매출 업데이트
                    print(f"{i} 붕어빵 {order[i]}개")
                    amount_sum += order[i]
                    total_payment += self.menu[i] * order[i]
                
                # 주문했을 때만 주문 개수 출력
                if amount_sum != 0:
                    print(f"총 {amount_sum}개 주문하셨습니다!")
                    print(f"계산하실 금액은 총 {total_payment}원입니다! :b")
                
                print("붕어빵 재고")
                for i in self.stocks:
                    print(f"{i}붕어빵 : {self.stocks[i]}개")

                print("-------------------------")
                break

            if self.stocks[self.fish[fish_type-1]] == 0:
                print("재고가 부족하여 판매할 수 없습니다!")
                continue

            while True:
                fish_count = input("붕어빵의 개수를 입력하세요!  ").strip()
                # 숫자 입력시에만 형변환
                if not(fish_count.isdigit()):
                    print("숫자만 입력해주세요!")
                    continue
                else:
                    fish_count = int(fish_count)

                # 재고 여부 확인 후 판매
                if self.stocks[self.fish[fish_type-1]] < fish_count:
                    print("재고가 부족하여 판매할 수 없습니다!")
                    print(f"현재 {self.fish[fish_type-1]}는 {self.stocks[self.fish[fish_type-1]]}개 있습니다! 다시 입력해주세요! :b")
                    continue

                order.update({self.fish[fish_type-1]:fish_count})
                
                print("현재주문:", end=" ")
                for i in order:
                    print(f"{i} 붕어빵 {order[i]}개", end=" ")
                print()
                break
```

```python
def admin_mode(self):
        print("관리자 모드")

        while True:
            func_type = input("실행할 기능을 선택하세요. 1.재고확인 2.매출확인 3.재고추가 4.종료").strip()
            # 숫자 입력시에만 형변환
            if not(func_type.isdigit()):
                print("숫자만 입력해주세요!")
                continue
            else:
                func_type = int(func_type)
            
            if func_type == 4:
                print("관리자 모드를 종료합니다.")
                print("-------------------------")
                break

            if func_type == 1:
                print("붕어빵 재고")
                for i in self.stocks:
                    print(f"{i}붕어빵 : {self.stocks[i]}개")
                print("-------------------------")
                continue

            if func_type == 2:
                total_sold = 0
                total_revenue = 0
                for i in self.sold_fish:
                    print(f"{i}붕어빵 : {self.sold_fish[i]}개")
                    total_sold += self.sold_fish[i]
                    total_revenue += self.menu[i] * self.sold_fish[i]
                print(f"총 {total_sold}개 판매 매출 금액 : {total_revenue}원")
                print("-------------------------")
                continue

            if func_type == 3:

                # 재고추가 초기화
                add_stocks = {}

                while True:
                    fish_type = input("추가할 재고를 입력하세요. 1.팥 2.슈크림 3.초코 4.종료 (중복 추가시 기존 추가 목록을 덮어씌움)").strip()
                    # 숫자 입력시에만 형변환
                    if not(fish_type.isdigit()):
                        print("숫자만 입력해주세요!")
                        continue
                    else:
                        fish_type = int(fish_type)
                        
                    # 기능 범위를 벗어나면 경고
                    if fish_type > (len(self.fish)+1) or fish_type < 1:
                        # print(type(fish_type))
                        print("1부터 4까지의 숫자만 입력해주세요! :b")
                        continue

                    # 숫자 4 입력시 종료
                    if fish_type == 4:
                        amount_sum = 0
                        for i in add_stocks:
                            self.stocks.update({ i : (self.stocks[i] + add_stocks[i]) })
                            print(f"{i} 붕어빵 {add_stocks[i]}개")
                            amount_sum += add_stocks[i]
                        
                        # 재고를 추가했을 때만 총 개수 출력
                        if amount_sum != 0:
                            print(f"총 {amount_sum}개 추가하셨습니다! :b")
                        
                        print("붕어빵 재고")
                        for i in self.stocks:
                            print(f"{i}붕어빵 : {self.stocks[i]}개")

                        print("-------------------------")
                        break

                    while True:
                        fish_count = input("붕어빵의 개수를 입력하세요!  ").strip()
                        # 숫자 입력시에만 형변환
                        if not(fish_count.isdigit()):
                            print("숫자만 입력해주세요!")
                            continue
                        else:
                            fish_count = int(fish_count)

                        add_stocks.update({self.fish[fish_type-1]:fish_count})
                        
                        print("현재 추가한 재고:", end=" ")
                        for i in add_stocks:
                            print(f"{i} 붕어빵 {add_stocks[i]}개", end=" ")
                        print()
                        break
```

설명: self.sales와 self.prices를 사용하여 총 매출을 계산하고 출력하세요.

```python
def calculate_total_sales(self):
        print("붕어빵 매출")
        total_sold = 0
        total_revenue = 0
        for i in self.sold_fish:
            print(f"{i}붕어빵 : {self.sold_fish[i]}개")
            total_sold += self.sold_fish[i]
            total_revenue += self.menu[i] * self.sold_fish[i]
        print(f"총 {total_sold}개 판매 매출 금액 : {total_revenue}원")

        print("-------------------------")
```
