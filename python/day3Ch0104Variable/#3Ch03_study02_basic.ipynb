코드 블럭 (for, if 등) 사용 시 하위 요소는 들여쓰기로 구분

예)

```python
i=1;
for 동물 in ("토끼", "사자", "푸바오"):
    print(동물)
    print("보러 갈래요?")
    i=i+1
    print(i)
print("###############")
for 동물 in ("토끼", "사자", "푸바오"):
    print(동물)
print("보러 갈래요?")
  # print("보러 갈래요?")
# 아무이유없이 들여쓰기 하면 에러
# 공백 2개는 탭 한번이랑 동일
```
