#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

10번만에 맞히지 못하면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

##본격적인 행맨 게임 규칙 설명

- words 리스트에 추가된 단어들을 사용해주세요
  - words = ["apple", "banana", "orange", "grape", "lemon"]

- 단어의 길이에 맞게 밑줄을 출력해주세요  
  (예)banana의 경우 _ _ _ _ _ _

- 사용자부터 1글자씩 입력을 받되,
   - 단어에 입력값이 포함되면 "단어를 맞추셨군요. 더 힘내봐요"라고 출력해주세요"
   - 단어에 입력값이 포함되어 있지 않다면 "틀렸습니다. 남은 시도 횟수 "란 메시지와 함꼐 남은 횟수를 출력해주세요

- 매번 입력을 받을 때마다 현재까지 맞힌 글자들 표시(맞히면 글자는 밑줄 출력)
    - 예) a 입력 시 : a _ _ _ _
          p 입력 시 : a p p _ _
          c 입력 시 : a p p _ _

- 모든 단어를 맞췄을 경우 "축하합니다! 단어를 맞추셨습니다"와 시도 횟수를 출력해주세요


행맨은 정말 유명 게임으로 찾아볼 수 있는 자료들이 많습니다.

```python
# 기본 라이브러리라 파이썬 설치시 자동으로 설치됩니다!
# random 모듈은 랜덤한 숫자를 생성할 때 사용합니다.
import random

# 게임에 사용될 단어 목록
words = ["apple", "banana", "orange", "grape", "lemon"]

# 행맨 그림
# 이거 먼가 했는데 남은 도전 횟수고 다 사용하면 목 매단 사람이 됨! ㄷㄷ
# 도전 횟수 HANGMAN!!
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]


class HangmanGame:

    # 클래스의 생성자 : 객체가 생성될 때 자동으로 호출되어 객체를 초기화하는 역할
    def __init__(self):

        # 중요한 목숨 카운트
        self.deadly_count = 0

    # 실시간 상황을 그림으로 표현하는 메소드
    def display(self):
        if self.deadly_count > 3:
            print(hangman_pics[self.deadly_count - 4])

    # main 메소드
    def play(self):

        input("안녕하세요!")

        # 게임을 시작하지 :b
        while True:
            he_want = input("행맨 게임을 시작하시겠습니까??? say Yes").strip()
            he_want = he_want.replace('아니', 'no').replace('싫', 'no').replace('안', 'no').replace('ㅜㅐ','no').replace('그만','no')

            # 게임 멈추는 명령어
            if he_want == '살려줘' or he_want == 'tkffuwnj':
                break

            # 싫다고 하면....
            if 'no' in he_want.lower():
                self.deadly_count = 10
                self.display()
                input("...")
                continue
            # 퀴즈 시작!
            else:
                # 랜덤 퀴즈 생성
                quiz = words[random.randint(0,len(words)-1)].strip()
                print(quiz)
                deadly_word = ''
                for i in quiz:
                    deadly_word += '_ '
                print(f"{len(quiz)}자리 영어 단어 {deadly_word}를 맞춰주세요")
                print("알파벳을 하나씩 추측해서 10번만에 맞히지 못하면 게임이 끝납니다.")
                
                # 맞춘 정답 리스트에 저장
                answers = []
                while True:

                    answer = input("한글자씩 입력하세요.").strip()

                    # 게임 멈추는 명령어
                    if answer == '살려줘ㅜㅜ' or answer == 'tkffuwnjnn':
                        break

                    # 두글자 이상 입력시 다시
                    if len(answer) > 1:
                        continue

                    # 빈값이나 공백 입력시 다시 (조건 안주면 맞았다고 함...)
                    # if answer in {'', ' '}:
                    if answer in {''}: # 빈값 입력시.  strip() 사용시 이것만 해도 됨.
                        continue
                    
                    print(f"입력한 글자는 {answer}입니다.")
                    input(f"입력한 글자는 {answer}입니다.")

                    # 시도 할 때 마다 남은 횟수 카운트
                    self.deadly_count += 1

                    # 맞았을 때
                    if answer in quiz:
                        # 맞춘 정답 리스트에 추가
                        answers.append(answer)
                        print("맞았습니다!")

                        # 맞춘 정답 리스트와 정답을 비교해 맞춘 글자를 표시
                        deadly_word = ''
                        for i in quiz:
                            if i in answers:
                                deadly_word += i+' '
                            else:
                                deadly_word += '_ '

                        # 모두 맞추면 성공!
                        if not ('_' in deadly_word):
                            print("축하합니다!  모두 맞추셨습니다!")
                            self.deadly_count = 4
                            print("You survived!")
                            self.display()
                            break

                        print(f"{len(quiz)}자리 영어 단어 {deadly_word}를 맞춰주세요.  남은 횟수 : {10 - self.deadly_count}")
                        continue

                    # 틀렸을 때

                    # 10번째에 틀리면... 죽음!!
                    if self.deadly_count == 10:
                        print("틀렸습니다!!!")
                        print("YOU ARE DEAD")
                        self.display()
                        input("...")
                        break

                    # 횟수가 7회 이하로 남았을 때부터 틀렸을 때 그림 보여줌.
                    print(f"틀렸습니다!  남은 횟수 : {10 - self.deadly_count}")
                    self.display()

                    # end of quiz while:

                # 퀴즈 종료 조건 만족시 프로그램 종료
                break

                # end of else:

            # end of program while:
            
if __name__ == "__main__":
    game = HangmanGame()
    game.play()

    
words = ["apple", "banana", "orange", "grape", "lemon"]
```
