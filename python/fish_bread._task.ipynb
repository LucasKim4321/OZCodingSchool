```python
fish = ['팥','슈크림','초코']
menu = { "팥":1000, "슈크림":1200, "초코":1500 }
stocks = { "팥":10, "슈크림":30, "초코":40 }
sold_fish = { "팥":10, "슈크림":20, "초코":5 }

print("-------------------------")
print("붕어빵 가격")
for i in menu:
    print(f"{i}붕어빵 : {menu[i]}원")

print("-------------------------")
print("붕어빵 재고")
for i in stocks:
    print(f"{i}붕어빵 : {stocks[i]}개")

print("-------------------------")
print("붕어빵 매출")
total_sold = 0
total_revenue = 0
for i in sold_fish:
    print(f"{i}붕어빵 : {sold_fish[i]}개")
    total_sold += sold_fish[i]
    total_revenue += menu[i] * sold_fish[i]
print(f"총 {total_sold}개 판매 매출 금액 : {total_revenue}원")

print("-------------------------")
while True:
    they_want = input("안녕하세요! 주문 하시겠습니까?").strip()  # 키오스크 종료 번호 : 1234 관리자 모드 : admin
    if they_want == '1234':
        print("키오스크를 종료합니다!")
        break
    # if '1234' in they_want:
    #     break

    elif they_want == 'admin' or they_want == '관리자':
        print("관리자 모드")
        
        while True:
            func_type = input("실행할 기능을 선택하세요. 1.재고확인 2.매출확인 3.재고추가 4.종료").strip()
            # 숫자 입력시에만 형변환
            if not(func_type.isdigit()):
                print("숫자만 입력해주세요!")
                continue
            else:
                func_type = int(func_type)
            
            if func_type == 4:
                print("관리자 모드를 종료합니다.")
                print("-------------------------")
                break

            if func_type == 1:
                print("붕어빵 재고")
                for i in stocks:
                    print(f"{i}붕어빵 : {stocks[i]}개")
                print("-------------------------")
                continue

            if func_type == 2:
                total_sold = 0
                total_revenue = 0
                for i in sold_fish:
                    print(f"{i}붕어빵 : {sold_fish[i]}개")
                    total_sold += sold_fish[i]
                    total_revenue += menu[i] * sold_fish[i]
                print(f"총 {total_sold}개 판매 매출 금액 : {total_revenue}원")
                print("-------------------------")
                continue

            if func_type == 3:
                
                # 재고추가 초기화
                add_stocks = {}

                while True:
                    fish_type = input("추가할 재고를 입력하세요. 1.팥 2.슈크림 3.초코 4.종료 (중복 추가시 기존 추가 목록을 덮어씌움)").strip()
                    # 숫자 입력시에만 형변환
                    if not(fish_type.isdigit()):
                        print("숫자만 입력해주세요!")
                        continue
                    else:
                        fish_type = int(fish_type)
                        
                    # 기능 범위를 벗어나면 경고
                    if fish_type > (len(fish)+1) or fish_type < 1:
                        # print(type(fish_type))
                        print("1부터 4까지의 숫자만 입력해주세요! :b")
                        continue

                    # 숫자 4 입력시 종료
                    if fish_type == 4:
                        amount_sum = 0
                        for i in add_stocks:
                            stocks.update({ i : (stocks[i] + add_stocks[i]) })
                            print(f"{i} 붕어빵 {add_stocks[i]}개")
                            amount_sum += add_stocks[i]
                        
                        # 재고를 추가했을 때만 총 개수 출력
                        if amount_sum != 0:
                            print(f"총 {amount_sum}개 추가하셨습니다! :b")
                        
                        print("붕어빵 재고")
                        for i in stocks:
                            print(f"{i}붕어빵 : {stocks[i]}개")

                        print("-------------------------")
                        break

                    while True:
                        fish_count = input("붕어빵의 개수를 입력하세요!  ").strip()
                        # 숫자 입력시에만 형변환
                        if not(fish_count.isdigit()):
                            print("숫자만 입력해주세요!")
                            continue
                        else:
                            fish_count = int(fish_count)

                        add_stocks.update({fish[fish_type-1]:fish_count})
                        
                        print("현재 추가한 재고:", end=" ")
                        for i in add_stocks:
                            print(f"{i} 붕어빵 {add_stocks[i]}개", end=" ")
                        print()
                        break
    
    else:
        # 매 주문 마다 주문 초기화
        order = {}

        while True:
            fish_type = input("주문할 붕어빵의 종류를 입력하세요 1.팥 2.슈크림 3.초코 4.주문 종료 (중복 주문시 기존 주문을 덮어씌움)").strip()
            # 숫자 입력시에만 형변환
            if not(fish_type.isdigit()):
                print("숫자만 입력해주세요!")
                continue
            else:
                fish_type = int(fish_type)
                
            # 주문 기능 범위를 벗어나면 경고
            #if의 조건문에서 형변환이 계속 유지됨. int()형변환 뒤 문자형 숫자인지 검사하면 문자가 아니라고 에러가 남! 조건식만 그럼. 어이 x
            if fish_type > (len(fish)+1) or fish_type < 1:
                # print(type(fish_type))
                print("1부터 4까지의 숫자만 입력해주세요! :b")
                continue

            # 숫자 4 입력시 주문 종료
            if fish_type == 4:
                amount_sum = 0
                total_payment = 0
                for i in order:
                    stocks.update({ i : (stocks[i] - order[i]) })  # 재고 업데이트
                    sold_fish.update({ i : (sold_fish[i] + order[i])})  # 매출 업데이트
                    print(f"{i} 붕어빵 {order[i]}개")
                    amount_sum += order[i]
                    total_payment += menu[i] * order[i]
                
                # 주문했을 때만 주문 개수 출력
                if amount_sum != 0:
                    print(f"총 {amount_sum}개 주문하셨습니다!")
                    print(f"계산하실 금액은 총 {total_payment}원입니다! :b")
                
                print("붕어빵 재고")
                for i in stocks:
                    print(f"{i}붕어빵 : {stocks[i]}개")

                print("-------------------------")
                break

            if stocks[fish[fish_type-1]] == 0:
                print("재고가 부족하여 판매할 수 없습니다!")
                continue

            while True:
                fish_count = input("붕어빵의 개수를 입력하세요!  ").strip()
                # 숫자 입력시에만 형변환
                if not(fish_count.isdigit()):
                    print("숫자만 입력해주세요!")
                    continue
                else:
                    fish_count = int(fish_count)

                # 재고 여부 확인 후 판매
                if stocks[fish[fish_type-1]] < fish_count:
                    print("재고가 부족하여 판매할 수 없습니다!")
                    print(f"현재 {fish[fish_type-1]}는 {stocks[fish[fish_type-1]]}개 있습니다! 다시 입력해주세요! :b")
                    continue

                order.update({fish[fish_type-1]:fish_count})
                
                print("현재주문:", end=" ")
                for i in order:
                    print(f"{i} 붕어빵 {order[i]}개", end=" ")
                print()
                break
```

```python
#for문 사용 붕어빵 주문

# 가격 리스트
메뉴 = [ ["팥", 2000], ["슈크림", 2500], ["잡채", 3000] ]

# 주문
개수 = []
총_가격 = 0
for i in range(len(메뉴)):
    개수.append(int(input(f"주문할 {메뉴[i][0]} 붕어빵 개수를 입력하세요. (숫자만 입력하세요)  ")))
    총_가격 += 메뉴[i][1] * 개수[i]

총_개수 = sum(개수)

# 결과
print("주문이 완료되었습니다!")
for i in range(len(메뉴)):
    print(f"{메뉴[i][0]} 붕어빵 {개수[i]}개", end = " ")

print(f"\n총 {총_개수}개 결제 금액은 {총_가격}원 입니다 :b")

# 추후 메뉴 변경 시 가격 리스트만 변경하면 됨! :b
```
