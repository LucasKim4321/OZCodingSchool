requests 모듈을 사용하여, 웹 페이지의 HTML을 가져오는 간단한 웹 스크레이퍼를 작성하세요.

'https://kream.co.kr/'에서<br> 
데이터를 가져와서 상태 코드와 HTML의 첫 100자를 출력하세요.<br>
(실습 전 requests 모듈을 설치해야 함)

```python
import requests

# 스크레이핑할 URL
url = 'https://kream.co.kr/'

# User-Agent 헤더 추가 # gpt왈 User-Agent 정보가 없으면 요청을 거절할 수도 있다함. 그래도 안됨..
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    # 웹 페이지 요청
    response = requests.get(url, headers=headers)
    
    # 상태 코드 출력
    print(f"Status Code: {response.status_code}")
    
    # HTML의 첫 100자 출력
    if response.status_code == 200:  # 요청 성공 시
        print("HTML 첫 100자:")
        print(response.text[:100])
    else:
        print("요청에 실패했습니다.")
except requests.exceptions.RequestException as e:
    print(f"오류 발생: {e}")
```

```python
# urllib.request 모듈을 임포트합니다. 이 모듈은 URL을 열고 읽는 기능을 제공합니다.
from urllib import request

# 지정된 URL("https://www.naver.com")을 엽니다.
target = request.urlopen("https://www.naver.com")
print(target)

# URL에서 읽은 웹 페이지의 내용을 변수에 저장합니다.
web_code = target.read()

# 웹 페이지의 내용을 출력합니다.
print(web_code)  # 예시: b'<!doctype html><html lang="ko">...</html>'
```
