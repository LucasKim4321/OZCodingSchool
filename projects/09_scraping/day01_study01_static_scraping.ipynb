```python
# 정적 스크랩핑

#  가상 환경 만들어서 requests 설치

# 가상환경 생성
# python3 -m venv .venv

# 가상환경 활성화
# source .venv/bin/activate

# 가상환경에서 모듈 설치
# pip3 install requests

# 커널 변경
# 우측 상단에 커널 클릭해서 가상환경으로 변경

import requests
from bs4 import BeautifulSoup

# 네이버 블로그 검색 정적 스크랩핑

base_url="https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query="
keyword = input("검색할 키워드를 입력해주세요~ ").strip()
url = base_url + keyword

req = requests.get(url)  # url -> "" -> 접속 가능한 텍스트 자료형 주소
# print(req)  # req 응답 결과가 코드로 반환됨. 정상 접속시 200
# print(req.text)  # req.text 접속한 웹페이지의 html코드를 텍스트로 반환
html = req.text

# beautifulsoup4 설치  https://www.crummy.com/software/BeautifulSoup/bs4/doc.ko/
# pip3 install beautifulsoup4

soup = BeautifulSoup(html, "html.parser")   # 가져온 text정보를 html태그로 인식
# print(soup)

# select 해당하는 모든 정보를 가져옵니다.
# select_one 해당하는 것중에서 가장 위에 있는 하나만 가져옵니다. 
results = soup.select(".view_wrap")  # view_wrap 클래스 가져오기. 태그로 인식
# print(results)

for rank, i in enumerate(results, 1):
    title = i.select_one(".title_link").text  # title_link클래스 가져온 후 텍스트를 가져옴
    writer = i.select_one(".name").text
    dsc = i.select_one(".dsc_link").text
    link = i.select_one(".title_link")["href"]

    print("-------")
    print(rank) # n번째 블로그
    print(f"제목: {title}")
    print(f"작성자: {writer}")
    print(f"글요약: {dsc}")
    print(f"링크: {link}")
    print("-------")
```
