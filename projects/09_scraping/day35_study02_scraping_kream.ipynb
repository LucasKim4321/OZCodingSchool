```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup

# URL 설정
base_url = "https://kream.co.kr/search?tab=products&keyword="
# keyword = input("키워드를 입력해주세요").strip()
keyword = "슈프림"
url = base_url + keyword

# User-Agent 설정
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"

# 옵션 설정
options_ = Options()
# User-Agent={} 형식이 아니라, --user-agent=...로 정확히 설정해야함.
options_.add_argument(f"--user-agent={user_agent}")  # --user-agent 형식으로 설정
# options_.add_experimental_option("detach", True)  # 브라우저 자동 종료 방지.  # 기본 적용된거 같은데.
options_.add_experimental_option("excludeSwitches", ["enable-logging", "enable-automation"])  # 따로 설정하면 마지막거만 설정됨. 이렇게 같이 적용

# enable-logging
# ChromeDriver가 실행될 때 터미널 또는 콘솔에 불필요한 디버그 로그 메시지를 출력하는 것을 방지합니다.
# 특히 DevTools 관련 불필요한 메시지가 출력되는 것을 막습니다.

# enable-automation
# Chrome을 자동화 모드(Selenium을 통한 실행)로 실행할 때 **"Chrome이 자동화된 테스트 소프트웨어에 의해 제어되고 있습니다."**라는 메시지를 제거합니다.

driver = webdriver.Chrome(options=options_)

# 적용된 옵션 확인
print(driver.capabilities)

# 웹페이지 접속
driver.get(url)


for i in range(3):

    # 0 부터 끝까지 스크로를 내리게 설정
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)

html = driver.page_source
# print(html)

soup = BeautifulSoup(html, "html.parser")   # 가져온 text정보를 html태그로 인식
# print(soup)

# select 해당하는 모든 정보를 가져옵니다.
# select_one 해당하는 것중에서 가장 위에 있는 하나만 가져옵니다. 
products = soup.select(".product_card")  #list 형태. CSS선택자 형식으로 가져옴.
# print(results)

for rank, i in enumerate(products, 1):
    brand = i.select_one(".product_info_brand, .brand").text  # title_link클래스 가져온 후 텍스트를 가져옴
    name_en = i.select_one(".product_info_product_name").text
    name_ko = i.select_one(".translated_name").text
    link = i.select_one("a")["href"]

    print("-------")
    print(rank) # n번째
    print(f"brand: {brand}")
    print(f"name_en: {name_en}")
    print(f"name_ko: {name_ko}")
    print(f"link: {link}")
    print("-------")

```
